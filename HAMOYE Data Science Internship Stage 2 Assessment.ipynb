```python
#importing libraries
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

Number 12: 
From the dataset, fit a linear model on the relationship between the temperature in the living room
in Celsius (x = T2) and the temperature outside the building (y = T6).
What is the R^2 value in two d.p.?

```python
x=df.T2.values.reshape(-1,1)
y=df.T6
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
LR=LinearRegression().fit(x_train,y_train)
y_pred=LR.predict(x_test)
```

```python
print('Coefficient of determination (R^2): %.2f'
      % r2_score(y_test, y_pred))
```

Number 13:
Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. 
The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility).
Run a multiple linear regression using the training set and evaluate your model on the test set. 
Answer the following questions:

What is the Mean Absolute Error (in two decimal places)?


```python
new_df = df.drop(['date', 'lights'], axis=1)
new_df.head()
```

```python
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(new_df), columns=new_df.columns)

features = normalized_df.drop(columns='Appliances')
target = normalized_df['Appliances']
```

```python
x_tra,x_tes,y_tra,y_tes=train_test_split(features,target,test_size=0.3,random_state=42)
```

```python
linear= LinearRegression()
linear.fit(x_tra, y_tra)
```

```python
pred = linear.predict(x_tes)
MAE = mean_absolute_error(y_test,pred)
print(f'MAE: {round(MAE,2)}')
```

Number 14:
What is the Residual Sum of Squares (in two decimal places)?

```python
rss = np.sum(np.square(y_tes - pred))
print(f'Residual Sum of Squares: {round(rss, 2)}')
```

Number 15:
What is the Root Mean Squared Error (in three decimal places)?

```python
rmse = np.sqrt(mean_squared_error(y_tes, pred))
print(f'RMSE: {round(rmse, 3)}')
```

Number 16
What is the Coefficient of Determination (in two decimal places)?

```python
print('Coefficient of determination (R^2): %.2f'
      % r2_score(y_tes,pred))
```

Number 17:
Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
def feature_weights(model, features):
    weights = pd.Series(model.coef_, features).sort_values()
    weights_df = pd.DataFrame(weights, columns=['Weight'])
    return weights_df


feature = features.columns
feature_weights( linear, feature)
```

Number 18:
Train a ridge regression model with an alpha value of 0.4.
Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_tra, y_tra)
ridge_pred = ridge_reg.predict(x_tes)

#obtaining the RMSE score
rmse = np.sqrt(mean_squared_error(y_tes, ridge_pred))
round(rmse, 3)
```

Number 19:
Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it.
How many of the features have non-zero feature weights?

```python
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_tra, y_tra)

feature_weights(lasso_reg, feature)
```

Number 20:
What is the new RMSE with the lasso regression? (Answer should be in three (3) decimal places)

```python
lasso_pred=lasso_reg.predict(x_tes)
rmse=np.sqrt(mean_squared_error(y_tes,lasso_pred))
print(f'Lasso RMSE: {round(rmse,3)}')
```
